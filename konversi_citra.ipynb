```python
import cv2
import matplotlib.pyplot as plt

nama_gambar = 'rio.jpeg'

image_bgr = cv2.imread(nama_gambar)

if image_bgr is None:
    print(f"Error: Gambar '{nama_gambar}' tidak ditemukan!")
else:
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title('Citra Asli (RGB)')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Citra Grayscale')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

nama_gambar = 'rio.jpeg'

image_bgr = cv2.imread(nama_gambar)

if image_bgr is not None:
    gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

    _, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Citra Grayscale')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(binary_image, cmap='gray')
    plt.title('Citra Biner')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Gambar '{nama_gambar}' tidak ditemukan!")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

nama_gambar = 'rio.jpeg'
m = 3

image_bgr = cv2.imread(nama_gambar)

if image_bgr is not None:
    gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

    mbit_image = np.uint8((gray_image >> (8 - m)) << (8 - m))

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Citra Grayscale (8-bit)')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(mbit_image, cmap='gray', vmin=0, vmax=255)
    plt.title(f'Citra {m}-bit')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Gambar '{nama_gambar}' tidak ditemukan!")
```

```python
import cv2
import matplotlib.pyplot as plt

nama_gambar = 'rio.jpeg'
nilai_brightness = 130

image_bgr = cv2.imread(nama_gambar)

if image_bgr is not None:
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    bright_image = cv2.convertScaleAbs(image_rgb, alpha=1, beta=nilai_brightness)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title('Citra Asli')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(bright_image)
    plt.title(f'Citra Brightness ({nilai_brightness})')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Gambar '{nama_gambar}' tidak ditemukan!")
```

```python
import cv2
import matplotlib.pyplot as plt

nama_gambar = 'rio.jpeg'
nilai_contrast = 3

image_bgr = cv2.imread(nama_gambar)

if image_bgr is not None:
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    contrast_image = cv2.convertScaleAbs(image_rgb, alpha=nilai_contrast, beta=0)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title('Citra Asli')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(contrast_image)
    plt.title(f'Citra Contrast ({nilai_contrast})')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Gambar '{nama_gambar}' tidak ditemukan!")
```

```python
import cv2
import matplotlib.pyplot as plt

nama_gambar = 'rio.jpeg'

image_bgr = cv2.imread(nama_gambar)

if image_bgr is not None:
    gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Citra Grayscale')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.hist(gray_image.ravel(), bins=256, range=[0, 256], color='black')
    plt.title('Histogram Grayscale')
    plt.xlabel('Intensitas Piksel')
    plt.ylabel('Jumlah Piksel')

    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Gambar '{nama_gambar}' tidak ditemukan!")
```

```python
import cv2
import matplotlib.pyplot as plt

nama_gambar = 'rio.jpeg'

image_bgr = cv2.imread(nama_gambar)

if image_bgr is not None:
    gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

    equalized_image = cv2.equalizeHist(gray_image)

    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Citra Grayscale Asli')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.hist(gray_image.ravel(), bins=256, range=[0, 256], color='black')
    plt.title('Histogram Asli')

    plt.subplot(2, 2, 3)
    plt.imshow(equalized_image, cmap='gray')
    plt.title('Citra Hasil Equalization')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.hist(equalized_image.ravel(), bins=256, range=[0, 256], color='black')
    plt.title('Histogram Hasil Equalization')

    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Gambar '{nama_gambar}' tidak ditemukan!")
```
